```python
def capping_outliers(column):
    q1=df['Years at Company'].quantile(0.25)
    q3=df['Years at Company'].quantile(0.75)
    iqr=q3-q1
    upper=q3+1.5*iqr
    lower=q1-1.5*iqr
    if column > upper:
        return upper
    elif column < lower:
        return lower
    else:
        return column
```

```python
df['Years at Company']=df['Years at Company'].apply(capping_outliers)
```
